To synchronize google colab with kaggle

Step1: Download kaggle on google colab
Download your account API token from kaggle (go to your account ----> create API token) and save it to your PC in json format    

Step2: Upload your account api that is json file ino google colab

Step3: Make your own kaggle directory on google colab so that all the kaggle data can be stored there, copy your uploaded kaggle.json file into kaggle folder and change the permissions of the file

Step4: Download the dataset from kaggle

Step5: usually dataset downloaded from kaggle will be in compressed format, hence we need to unzip it
